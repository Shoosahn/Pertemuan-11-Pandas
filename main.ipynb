```python
import pandas as pd

# Load data
df = pd.read_csv(
        "D:\Kuliah\PEMR. DASAR\Pertemuan 11 - Pandas\data\disperkim-od_16984_jumlah_sampah_yang_ditangani_berdasarkan_kabupatenkota_data.csv"

)

# Kolom yang relevan
df = df[['nama_kabupaten_kota', 'jumlah_sampah', 'tahun']]

print(df.head())
```

```python
tahun_target = 2023
total_sampah = 0

for _, row in df.iterrows():
    if row['tahun'] == tahun_target:
        total_sampah += row['jumlah_sampah']

print(f"Total sampah ditangani di Jawa Barat tahun {tahun_target}: {total_sampah} ton")
```

```python
total_per_tahun = {}

for _, row in df.iterrows():
    tahun = row['tahun']
    total_per_tahun[tahun] = total_per_tahun.get(tahun, 0) + row['jumlah_sampah']

df_total_tahun = pd.DataFrame(
    list(total_per_tahun.items()),
    columns=['tahun', 'total_sampah']
)

pd.options.display.float_format = '{:,.0f}'.format
print(df_total_tahun)
```

```python
rekap = {}

for _, row in df.iterrows():
    key = (row['nama_kabupaten_kota'], row['tahun'])
    rekap[key] = rekap.get(key, 0) + row['jumlah_sampah']

df_rekap = pd.DataFrame(
    [(k[0], k[1], v) for k, v in rekap.items()],
    columns=['nama_kabupaten_kota', 'tahun', 'total_sampah']
)

print(df_rekap.head())
```

```python
df_total_tahun.to_csv("total_sampah_per_tahun.csv", index=False)
df_total_tahun.to_excel("total_sampah_per_tahun.xlsx", index=False)

df_rekap.to_csv("total_sampah_per_kabupaten_tahun.csv", index=False)
df_rekap.to_excel("total_sampah_per_kabupaten_tahun.xlsx", index=False)
```
